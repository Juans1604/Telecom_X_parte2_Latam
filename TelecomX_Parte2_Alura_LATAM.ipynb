#Extracción del archivo

```python
import pandas as pd

datos = pd.read_csv('/content/datos_normalizados.csv')
datos.head()
```

```python
datos.info()
```

#Eliminación de columnas irrelevantes

```python
datos = datos.drop(columns=["customerID","Cuentas_Diarias", "FacturaTotal"])

datos.head()
```

#Encoding

```python
datos_encoded = pd.get_dummies(datos, drop_first=True)

datos_encoded.head()
```

#Verificación de la proporción y Balanceo de clases

```python
from imblearn.over_sampling import SMOTE
# --- 4. Separar variables predictoras (X) y objetivo (y) ---
X = datos_encoded.drop(columns=["Cancelacion"])
y = datos_encoded["Cancelacion"]

# --- 5. Proporciones originales ---
print("Proporción original:")
print(y.value_counts(normalize=True))

# --- 6. Aplicar SMOTE ---
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# --- 7. Proporciones balanceadas ---
print("\nProporción después de aplicar SMOTE:")
print(y_res.value_counts(normalize=True))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr_cancelacion = datos_encoded.corr()["Cancelacion"].sort_values(ascending=False)

top_corr = corr_cancelacion[1:11]

plt.figure(figsize=(10, 6))
top_corr.plot(kind="bar", color="skyblue")
plt.title("Top 10 variables más correlacionadas con Cancelación")
plt.ylabel("Coeficiente de correlación")
plt.xlabel("Variables")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x="Cancelacion", y="MesesConEmpresa", data=datos_encoded)
plt.title("Tiempo de Contrato vs Cancelación")
plt.xlabel("Cancelación (0 = No, 1 = Sí)")
plt.ylabel("Tiempo de Contrato")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

plt.figure(figsize=(6, 4))
sns.boxplot(x="Cancelacion", y="FacturaMensual", data=datos_encoded)
plt.title("Gasto Total vs Cancelación")
plt.xlabel("Cancelación (0 = No, 1 = Sí)")
plt.ylabel("Gasto Total")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

plt.figure(figsize=(7, 5))
sns.scatterplot(x="MesesConEmpresa", y="FacturaMensual", hue="Cancelacion", data=datos_encoded, alpha=0.6)
plt.title("Tiempo de Contrato y Gasto Total vs Cancelación")
plt.xlabel("Tiempo de Contrato")
plt.ylabel("Gasto Total")
plt.legend(title="Cancelación", labels=["No", "Sí"])
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

#Separación de datos y Creación de modelos

```python
from sklearn.model_selection import train_test_split

# Separar variables predictoras (X) y variable objetivo (y)
X = datos_encoded.drop("Cancelacion", axis=1)
y = datos_encoded["Cancelacion"]

# División 70% entrenamiento - 30% prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
print("Tamaño del conjunto de entrenamiento:", X_train.shape)
print("Tamaño del conjunto de prueba:", X_test.shape)
print("\nProporción en entrenamiento:", y_train.value_counts(normalize=True))
print("\nProporción en prueba:", y_test.value_counts(normalize=True))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

log_model = LogisticRegression(max_iter=1000, random_state=42)
log_model.fit(X_train_scaled, y_train)
y_pred_log = log_model.predict(X_test_scaled)

print("Regresión Logística (con normalización)")
print("Accuracy:", accuracy_score(y_test, y_pred_log))
print(classification_report(y_test, y_pred_log))

tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_train, y_train)
y_pred_tree = tree_model.predict(X_test)

print("\nÁrbol de Decisión (sin normalización)")
print("Accuracy:", accuracy_score(y_test, y_pred_tree))
print(classification_report(y_test, y_pred_tree))

fig, axes = plt.subplots(1, 2, figsize=(12,5))

sns.heatmap(confusion_matrix(y_test, y_pred_log), annot=True, fmt="d", cmap="Blues", ax=axes[0])
axes[0].set_title("Matriz de Confusión - Regresión Logística")
axes[0].set_xlabel("Predicción")
axes[0].set_ylabel("Real")

sns.heatmap(confusion_matrix(y_test, y_pred_tree), annot=True, fmt="d", cmap="Greens", ax=axes[1])
axes[1].set_title("Matriz de Confusión - Árbol de Decisión")
axes[1].set_xlabel("Predicción")
axes[1].set_ylabel("Real")

plt.tight_layout()
plt.show()
```

##Explicacion de decisión


*   Regresión Logística (con normalización):
La normalización evita que variables con magnitudes grandes (ej. Gasto Total) dominen sobre otras más pequeñas. Así los coeficientes se calculan de manera justa.

*   Árbol de Decisión (sin normalización):
Este modelo divide los datos con reglas basadas en umbrales, por lo que la escala de las variables no afecta la calidad de las divisiones.



#Evaluación de los modelos

```python
from sklearn.metrics import precision_score, recall_score, f1_score

accuracy_log = accuracy_score(y_test, y_pred_log)
precision_log = precision_score(y_test, y_pred_log)
recall_log = recall_score(y_test, y_pred_log)
f1_log = f1_score(y_test, y_pred_log)

print("Regresión Logística (con normalización)")
print(f"Accuracy: {accuracy_log:.4f}")
print(f"Precisión: {precision_log:.4f}")
print(f"Recall: {recall_log:.4f}")
print(f"F1-score: {f1_log:.4f}")
print("\nReporte de Clasificación:\n", classification_report(y_test, y_pred_log))

accuracy_tree = accuracy_score(y_test, y_pred_tree)
precision_tree = precision_score(y_test, y_pred_tree)
recall_tree = recall_score(y_test, y_pred_tree)
f1_tree = f1_score(y_test, y_pred_tree)

print("\nÁrbol de Decisión (sin normalización)")
print(f"Accuracy: {accuracy_tree:.4f}")
print(f"Precisión: {precision_tree:.4f}")
print(f"Recall: {recall_tree:.4f}")
print(f"F1-score: {f1_tree:.4f}")
print("\nReporte de Clasificación:\n", classification_report(y_test, y_pred_tree))

fig, axes = plt.subplots(1, 2, figsize=(12,5))

sns.heatmap(confusion_matrix(y_test, y_pred_log), annot=True, fmt="d", cmap="Blues", ax=axes[0])
axes[0].set_title("Matriz de Confusión - Regresión Logística")
axes[0].set_xlabel("Predicción")
axes[0].set_ylabel("Real")

sns.heatmap(confusion_matrix(y_test, y_pred_tree), annot=True, fmt="d", cmap="Greens", ax=axes[1])
axes[1].set_title("Matriz de Confusión - Árbol de Decisión")
axes[1].set_xlabel("Predicción")
axes[1].set_ylabel("Real")

plt.tight_layout()
plt.show()
```

##Comparación critica

Análisis crítico:
- La Regresión Logística mostró mejor exactitud global.
- La Regresión Logística identificó mejor a los clientes que cancelan (mayor recall).
- La Regresión Logística cometió menos falsos positivos (mejor precisión).

Posibles problemas:
- Si el Árbol de Decisión tiene accuracy muy alto en entrenamiento pero bajo en test → Overfitting.
- Si ambos modelos tienen resultados bajos en todas las métricas → Underfitting (necesitan más ajuste o características).
- Una estrategia para mejorar el Árbol de Decisión sería limitar su profundidad (max_depth).
- Una estrategia para mejorar la Regresión Logística sería ajustar regularización (C, penalty).

#Análisis

##1. Regresión Logística

En este modelo, cada variable recibe un coeficiente que indica su influencia en la probabilidad de que un cliente cancele:

Si el coeficiente es positivo, significa que cuando esa variable aumenta, también aumenta la probabilidad de cancelación.

Si el coeficiente es negativo, significa que cuando esa variable aumenta, la probabilidad de cancelación disminuye.

La magnitud (el valor absoluto del coeficiente) indica qué tan fuerte es ese efecto.

Ejemplo:

Si la variable Tiempo de contrato tiene un coeficiente negativo, quiere decir que a mayor tiempo de contrato, menor probabilidad de cancelación.

Si la variable Gasto total tiene un coeficiente positivo y grande, entonces los clientes con mayor gasto total tienden a cancelar más.

##2. Random Forest

Este modelo no utiliza coeficientes, sino que calcula la importancia de cada variable con base en cuánto contribuye a mejorar las divisiones en los árboles:

Una variable es importante si aparece muchas veces en los nodos de los árboles y ayuda a separar bien a los clientes que cancelan de los que no cancelan.

La importancia se mide en una escala relativa (ej: 0.30 significa que esa variable explicó el 30% de las divisiones más útiles).

Ejemplo:

Si Tiempo de contrato aparece como la variable más importante (ej. 0.25), quiere decir que es el principal factor para decidir si un cliente cancela.

Si Método de pago tiene una importancia baja (ej. 0.05), entonces influye mucho menos en la decisión del modelo.

3. **Método de pago** (débito automático está asociado con menor churn).  
4. **Contratación de servicios adicionales** (quienes contratan más de un servicio permanecen más tiempo).  

---

## 6. Estrategias de Retención

### Recomendaciones:
